## <font color='red'>Fill Nan values<font>

```python
cat_feature=list(train.select_dtypes(include='object').columns)
```

```python
print(cat_feature,end='')
```

```python
len(cat_feature)
```

```python
num_feat=list(train.select_dtypes(exclude='object').columns)
```

```python
len(num_feat)
```

```python
def fill_missing(df):#pass by reference 
    for i in cat_feature:
        df[i].fillna(value=df[i].mode()[0],inplace=True)
    for i in num_feat:
        df[i].fillna(value=df[i].mean(),inplace=True)
```

```python
fill_missing(train)
```
